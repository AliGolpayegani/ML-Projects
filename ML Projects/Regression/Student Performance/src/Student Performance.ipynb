```python
X = df.drop(["Performance Index"], axis=1)
y = df["Performance Index"].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
scaler = MinMaxScaler()

y_reshape = np.reshape(y, (-1, 1))
print(y_reshape)
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y_reshape)
```

```python
X_train, X_validation, y_train, y_validation = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=1234)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
r2 = r2_score(y_validation, model.predict(X_validation))
mse = mean_squared_error(y_validation, model.predict(X_validation))
print(f"R2 Score: {r2}")
print(f"MSE: {mse}")
```
